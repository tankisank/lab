```python
def float_format(vector, decimal):
    return np.round((vector).astype(float), decimals=decimal)
```

```python
from fractions import Fraction
import numpy as np
```

```python
#(defaultProbability)
dp = Fraction(1,4)


# WWW matrix
M = np.matrix([[0, Fraction(1, 2), 0, 0],
    [Fraction(1, 3), 0, 0, Fraction(1, 2)],
    [Fraction(1, 3), 0, 0, Fraction(1, 2)],
    [Fraction(1, 3), Fraction(1, 2), 0, 0]])


E = np.zeros((4,4))
E[:] = dp

# random walk parameter 
beta = 0.85

# WWW matrix
A = beta * M + ((1-beta) * E)

# initial vector
r = np.matrix([dp, dp, dp,dp])
r = np.transpose(r)

previous_r = r
for it in range(1,100):
    r = A * r
    print (float_format(r,3))
    #check if converged
    if (previous_r==r).all():
        break
    previous_r = r

print ("Final:\n", float_format(r,3))    
print ("sum", np.sum(r))
    
## example 1.2
print ("PageRank*3:", np.round((3*r).astype(np.float), decimals=3))
```
